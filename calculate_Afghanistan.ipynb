```python
import pandas as pd 
import numpy as np 

data = pd.read_csv("Life Expectancy Data.csv")
afghan_data = data[data["Country"]== "Afghanistan"]
train_data = afghan_data[2:]
test_data = afghan_data[:2]
X_train = train_data.iloc[:, [16]]
y_train = train_data.iloc[:,[3]]
X_test = test_data.iloc[:, [16]]
y_test = test_data.iloc[:,[3]]

#print(X.shape, y.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

import math


#Creates the polynomial Data Sets for Test and Train
polynomialDataTest = X_test.to_numpy()
polyDataTest = polynomialDataTest.reshape(-1,1)
quadratic_feature = PolynomialFeatures(degree = 2)
X_quad_test = quadratic_feature.fit_transform(polyDataTest)


polynomialDataTrain = X_train.to_numpy()
polyDataTrain = polynomialDataTrain.reshape(-1,1)
quadratic_feature = PolynomialFeatures(degree = 2)
X_quad_train = quadratic_feature.fit_transform(polyDataTrain)

#Cubic
cubic_feature = PolynomialFeatures(degree = 3)
#Cubic test
X_cubic_test = cubic_feature.fit_transform(polyDataTest)
#Cubic train
X_cubic_train = cubic_feature.fit_transform(polyDataTrain)

#Quartic
quartic_feature = PolynomialFeatures(degree = 4)
#Cubic test
X_quartic_test = quartic_feature.fit_transform(polyDataTest)
#Cubic train
X_quartic_train = quartic_feature.fit_transform(polyDataTrain)


#Creates the actual models for all types
#Lin Reg model
linear_regression = LinearRegression().fit(X_train,y_train)
#Quad Reg Model
quad_reg = LinearRegression(fit_intercept=False).fit(X_quad_train, y_train)
#Cubic Reg Model
cubic_reg = LinearRegression(fit_intercept=False).fit(X_cubic_train,y_train)
#Quartic Reg Model
quartic_reg = LinearRegression(fit_intercept=False).fit(X_quartic_train,y_train)


def rmse(y_true, y_pred):
    mse = sum((y_true - y_pred)**2) / len(y_true)
    return math.sqrt(mse)


linRegPred_TEST = linear_regression.predict(X_test)
linRegReal_TEST = y_test.to_numpy()

linRegPred_TRAIN = linear_regression.predict(X_train)
linRegReal_TRAIN = y_train.to_numpy()

#Quad MAKING PREDICTIONS and setting UP y_test as an array for both test + train
quad_regPred_TEST = quad_reg.predict(X_quad_test) #for test its not X_quad
quad_regReal_TEST = y_test.to_numpy()

quad_regPred_TRAIN = quad_reg.predict(X_quad_train)
quad_regReal_TRAIN = y_train.to_numpy()

#CUBIC MAKING PREDICTIONS and setting UP y_test as an array for both test + train
cubic_regPred_TEST = cubic_reg.predict(X_cubic_test) 
cubic_regReal_TEST = y_test.to_numpy()

cubic_regPred_TRAIN = cubic_reg.predict(X_cubic_train)
cubic_regReal_TRAIN = y_train.to_numpy()

#Quartic MAKING PREDICTIONS and setting UP y_test as an array for both test + train
quartic_regPred_TEST = quartic_reg.predict(X_quartic_test) 
quartic_regReal_TEST = y_test.to_numpy()

quartic_regPred_TRAIN = quartic_reg.predict(X_quartic_train)
quartic_regReal_TRAIN = y_train.to_numpy()




print("Lin Reg TEST RMSE " + str(rmse(linRegPred_TEST, linRegReal_TEST)))
print("Lin Reg TRAIN RMSE " + str(rmse(linRegPred_TRAIN, linRegReal_TRAIN)))

print("Quad Reg TEST RMSE " + str(rmse(quad_regPred_TEST, quad_regReal_TEST)))
print("Quad Reg TRAIN RMSE " + str(rmse(quad_regPred_TRAIN, quad_regReal_TRAIN)))

print("Cubic Reg TEST RMSE " + str(rmse(cubic_regPred_TEST, cubic_regReal_TEST)))
print("Cubic Reg TRAIN RMSE " + str(rmse(cubic_regPred_TRAIN, cubic_regReal_TRAIN)))

print("Quartic Reg TEST RMSE " + str(rmse(quartic_regPred_TEST, quartic_regReal_TEST)))
print("Quartic Reg TRAIN RMSE " +str(rmse(quartic_regPred_TRAIN, quartic_regReal_TRAIN)))

print("Lin Reg TEST R^2 " + str(r2_score(linRegReal_TEST, linRegPred_TEST)))
print("Lin Reg TRAIN R^2 " + str(r2_score(linRegReal_TRAIN, linRegPred_TRAIN)))

print("Quad Reg TEST R^2 " + str(r2_score(quad_regReal_TEST, quad_regPred_TEST)))
print("Quad Reg TRAIN R^2 " + str(r2_score(quad_regReal_TRAIN, quad_regPred_TRAIN)))

print("Cubic Reg TEST R^2 " + str(r2_score(cubic_regReal_TEST, cubic_regPred_TEST)))
print("Cubic Reg TRAIN R^2 " + str(r2_score(cubic_regReal_TRAIN, cubic_regPred_TRAIN,)))

print("Quartic Reg TEST R^2 " + str(r2_score(quartic_regReal_TEST, quartic_regPred_TEST)))
print("Quartic Reg TRAIN R^2 " + str(r2_score( quartic_regReal_TRAIN, quartic_regPred_TRAIN)))

```
