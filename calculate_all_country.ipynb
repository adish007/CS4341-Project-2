```python
import pandas as pd 
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

import math


data = pd.read_csv("Life Expectancy Data.csv").dropna(subset=["GDP"])

def rmse(y_true, y_pred):
    mse = sum((y_true - y_pred)**2) / len(y_true)
    return math.sqrt(mse)

developing = data[data["Status"]=="Developing"]
developed = data[data["Status"]=="Developed"]



developedCountriesList = list(developed['Country'].unique()) # gets unique developed countries
developingCountriesList = list(developing['Country'].unique()) #gets unique developing countries


# print( developedCountriesList)
# print(developingCountriesList)


# Now work for only developed stuff 
#LINEAR
developed_countries_TEST_LIN = [] # stores the real rmse of each country
developed_countries_TRAIN_LIN = [] # stores the real rmse of each country 
#QUADRATIC
developed_countries_TEST_QUAD = [] # stores the real rmse of each country
developed_countries_TRAIN_QUAD = [] # stores the real rmse of each country 
#CUBIC
developed_countries_TEST_CUB = [] # stores the real rmse of each country
developed_countries_TRAIN_CUB = [] # stores the real rmse of each country 
#QUARTIC
developed_countries_TEST_QUAR = [] # stores the real rmse of each country
developed_countries_TRAIN_QUAR = [] # stores the real rmse of each country 

#LINEAR
developed_countries_TEST_LIN_R2 = [] # stores the r2 of each country
developed_countries_TRAIN_LIN_R2 = [] # stores the r2 of each country 
#QUADRATIC
developed_countries_TEST_QUAD_R2 = [] # stores the r2 of each country
developed_countries_TRAIN_QUAD_R2 = [] # stores the r2 of each country 
#CUBIC
developed_countries_TEST_CUB_R2 = [] # stores the r2 of each country 
developed_countries_TRAIN_CUB_R2 = [] # stores the r2 of each country 
#QUARTIC
developed_countries_TEST_QUAR_R2 = [] # stores the r2 of each country 
developed_countries_TRAIN_QUAR_R2 = [] # stores the r2 of each country 

for country in developedCountriesList:
    
    CountryData = data[data["Country"]== country] # creates the train and test for the current country
    train_data = CountryData[2:]
    test_data = CountryData[:2]
    X_train = train_data.iloc[:, [16]]
    y_train = train_data.iloc[:,[3]]
    X_test = test_data.iloc[:, [16]]
    y_test = test_data.iloc[:,[3]]

    #Creates the polynomial Data Sets for Test and Train
    polynomialDataTest = X_test.to_numpy()
    polyDataTest = polynomialDataTest.reshape(-1,1)
    polynomialDataTrain = X_train.to_numpy()
    polyDataTrain = polynomialDataTrain.reshape(-1,1)

    #Quadratic
    quadratic_feature = PolynomialFeatures(degree = 2)
    #Quad Test
    X_quad_test = quadratic_feature.fit_transform(polyDataTest)
    #Quad Train
    X_quad_train = quadratic_feature.fit_transform(polyDataTrain)

    #Cubic
    cubic_feature = PolynomialFeatures(degree = 3)
    #Cubic test
    X_cubic_test = cubic_feature.fit_transform(polyDataTest)
    #Cubic train
    X_cubic_train = cubic_feature.fit_transform(polyDataTrain)

    #Quartic
    quartic_feature = PolynomialFeatures(degree = 4)
    #Cubic test
    X_quartic_test = quartic_feature.fit_transform(polyDataTest)
    #Cubic train
    X_quartic_train = quartic_feature.fit_transform(polyDataTrain)



    linear_regression = LinearRegression().fit(X_train,y_train) # creates current country linReg model
    #Quad Reg Model
    quad_reg = LinearRegression(fit_intercept=False).fit(X_quad_train, y_train)
    #Cubic Reg Model
    cubic_reg = LinearRegression(fit_intercept=False).fit(X_cubic_train,y_train)
    #Quartic Reg Model
    quartic_reg = LinearRegression(fit_intercept=False).fit(X_quartic_train,y_train)


    linRegPred_TEST = linear_regression.predict(X_test)
    linRegReal_TEST = y_test.to_numpy()

    linRegPred_TRAIN = linear_regression.predict(X_train)
    linRegReal_TRAIN = y_train.to_numpy()
    #Quad MAKING PREDICTIONS and setting UP y_test as an array for both test + train
    quad_RegPred_TEST = quad_reg.predict(X_quad_test) #for test its not X_quad
    quad_RegReal_TEST = y_test.to_numpy()

    quad_RegPred_TRAIN = quad_reg.predict(X_quad_train)
    quad_RegReal_TRAIN = y_train.to_numpy()

    #CUBIC MAKING PREDICTIONS and setting UP y_test as an array for both test + train
    cubic_RegPred_TEST = cubic_reg.predict(X_cubic_test) 
    cubic_RegReal_TEST = y_test.to_numpy()

    cubic_RegPred_TRAIN = cubic_reg.predict(X_cubic_train)
    cubic_RegReal_TRAIN = y_train.to_numpy()

    #Quartic MAKING PREDICTIONS and setting UP y_test as an array for both test + train
    quartic_RegPred_TEST = quartic_reg.predict(X_quartic_test) 
    quartic_RegReal_TEST = y_test.to_numpy()

    quartic_RegPred_TRAIN = quartic_reg.predict(X_quartic_train)
    quartic_RegReal_TRAIN = y_train.to_numpy()  


    
    developed_countries_TEST_LIN.append(rmse(linRegPred_TEST, linRegReal_TEST))
    developed_countries_TRAIN_LIN.append(rmse(linRegPred_TRAIN, linRegReal_TRAIN))

    developed_countries_TEST_QUAD.append(rmse(quad_RegPred_TEST, quad_RegReal_TEST))
    developed_countries_TRAIN_QUAD.append(rmse(quad_RegPred_TRAIN, quad_RegReal_TRAIN))

    developed_countries_TEST_CUB.append(rmse(cubic_RegPred_TEST, cubic_RegReal_TEST))
    developed_countries_TRAIN_CUB.append(rmse(cubic_RegPred_TRAIN, cubic_RegReal_TRAIN))

    developed_countries_TEST_QUAR.append(rmse(quartic_RegPred_TEST, quartic_RegReal_TEST))
    developed_countries_TRAIN_QUAR.append(rmse(quartic_RegPred_TRAIN, quartic_RegReal_TRAIN))

    developed_countries_TEST_LIN_R2.append(r2_score(linRegReal_TEST,linRegPred_TEST))
    developed_countries_TRAIN_LIN_R2.append(r2_score(linRegReal_TRAIN,linRegPred_TRAIN))

    developed_countries_TEST_QUAD_R2.append(r2_score(quad_RegReal_TEST,quad_RegPred_TEST))
    developed_countries_TRAIN_QUAD_R2.append(r2_score(quad_RegReal_TRAIN,quad_RegPred_TRAIN))

    developed_countries_TEST_CUB_R2.append(r2_score(cubic_RegReal_TEST,cubic_RegPred_TEST))
    developed_countries_TRAIN_CUB_R2.append(r2_score(cubic_RegReal_TRAIN,cubic_RegPred_TRAIN))

    developed_countries_TEST_QUAR_R2.append(r2_score(quartic_RegReal_TEST,quartic_RegPred_TEST))
    developed_countries_TRAIN_QUAR_R2.append(r2_score(quartic_RegReal_TRAIN,quartic_RegPred_TRAIN))



    
print("Developed RMSE LINEAR TEST "+ str(sum(developed_countries_TEST_LIN)/len(developed_countries_TEST_LIN)))
print("Developed RMSE LINEAR TRAIN " + str(sum(developed_countries_TRAIN_LIN)/len(developed_countries_TRAIN_LIN)))

print("Developed RMSE QUADRATIC TEST "+ str(sum(developed_countries_TEST_QUAD)/len(developed_countries_TEST_QUAD)))
print("Developed RMSE QUADRATIC TRAIN " + str(sum(developed_countries_TRAIN_QUAD)/len(developed_countries_TRAIN_QUAD)))

print("Developed RMSE CUBIC TEST "+ str(sum(developed_countries_TEST_CUB)/len(developed_countries_TEST_CUB)))
print("Developed RMSE CUBIC TRAIN " + str(sum(developed_countries_TRAIN_CUB)/len(developed_countries_TRAIN_CUB)))

print("Developed RMSE QUARTIC TEST "+ str(sum(developed_countries_TEST_QUAR)/len(developed_countries_TEST_QUAR)))
print("Developed RMSE QUARTIC TRAIN " + str(sum(developed_countries_TRAIN_QUAR)/len(developed_countries_TRAIN_QUAR)))


print("Developed R2 LINEAR TEST "+ str(sum(developed_countries_TEST_LIN_R2)/len(developed_countries_TEST_LIN_R2)))
print("Developed R2 LINEAR TRAIN " + str(sum(developed_countries_TRAIN_LIN_R2)/len(developed_countries_TRAIN_LIN_R2)))

print("Developed R2 QUADRATIC TEST "+ str(sum(developed_countries_TEST_QUAD_R2)/len(developed_countries_TEST_QUAD_R2)))
print("Developed R2 QUADRATIC TRAIN " + str(sum(developed_countries_TRAIN_QUAD_R2)/len(developed_countries_TRAIN_QUAD_R2)))

print("Developed R2 CUBIC TEST "+ str(sum(developed_countries_TEST_CUB_R2)/len(developed_countries_TEST_CUB_R2)))
print("Developed R2 CUBIC TRAIN " + str(sum(developed_countries_TRAIN_CUB_R2)/len(developed_countries_TRAIN_CUB_R2)))


print("Developed R2 QUARTIC TEST "+ str(sum(developed_countries_TEST_QUAR_R2)/len(developed_countries_TEST_QUAR_R2)))
print("Developed R2 QUARTIC TRAIN " + str(sum(developed_countries_TRAIN_QUAR_R2)/len(developed_countries_TRAIN_QUAR_R2)))




# print("Developed R2 LINEAR TEST " + str(r2_score(linRegReal_TEST,linRegPred_TEST)))
# print("Developed R2 LINEAR TRAIN " + str(r2_score(linRegReal_TRAIN,linRegPred_TRAIN)))

# print("Developed R2 QUADRATIC TEST " + str(r2_score(quad_RegReal_TEST,quad_RegPred_TEST)))
# print("Developed R2 QUADRATIC TRAIN " + str(r2_score(quad_RegReal_TRAIN,quad_RegPred_TRAIN)))

# print("Developed R2 CUBIC TEST " + str(r2_score(cubic_RegReal_TEST,cubic_RegPred_TEST)))
# print("Developed R2 CUBIC TRAIN " + str(r2_score(cubic_RegReal_TRAIN,cubic_RegPred_TRAIN)))

# print("Developed R2 QUARTIC TEST " + str(r2_score(quartic_RegReal_TEST,quartic_RegPred_TEST)))
# print("Developed R2 QUARTIC TRAIN " + str(r2_score(quartic_RegReal_TRAIN,quartic_RegPred_TRAIN)))


 
```
